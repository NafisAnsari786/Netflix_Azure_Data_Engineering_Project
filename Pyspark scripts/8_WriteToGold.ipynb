## Writing data to Gold layer for backup


```python
# 1. Define the ADLS Gold container path
gold_container_path = "abfss://gold@netflixprojectnafis.dfs.core.windows.net/"

# 2. List of finalized Netflix Gold tables from the DLT run
gold_tables = [
    "gold_netflixcast",
    "gold_netflixcategory",
    "gold_netflixcountries",
    "gold_netflixdirectors",
    "gold_netflixtitles"
]

# 3. Loop through and write each table to the ADLS gold container
for table_name in gold_tables:
    print(f"Reading {table_name} from DLT Schema...")
    # Read the table from specific catalog and schema
    df = spark.read.table(f"netflix_catalog.dlt_schema.{table_name}")

    print(f"Writing {table_name} to ADLS Gold container...")
    # Write physically to ADLS as Delta format
    df.write.format("delta").mode("overwrite")\
        .save(f"{gold_container_path}{table_name}")
print("All Netflix Gold tables successfully written to Azure Data Lake !")
```
