# Silver Notebook Lookup Tables


### Parameters

```python
dbutils.widgets.text("sourcefolder", "netflix_directors")
dbutils.widgets.text("targetfolder", "netflix_directors")
```

### Variables

```python
var_src_folder = dbutils.widgets.get("sourcefolder")
var_trg_folder = dbutils.widgets.get("targetfolder")
```

```python
df = spark.read.format("csv")\
        .option("header", True)\
        .option("inferSchema", True)\
        .load(f"abfss://bronze@netflixprojectnafis.dfs.core.windows.net/{var_src_folder}")
        
```

```python
display(df)
```

```python
#write data

df.write.format("delta")\
    .mode("append")\
    .option("path", f"abfss://silver@netflixprojectnafis.dfs.core.windows.net/{var_trg_folder}")\
    .save()
```
